```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
file = Path("Cleaned_Data/merged_all_data.csv")

data = pd.read_csv(file)
df = pd.DataFrame(data)
#Changing index from default to postcode column
df.set_index("POST_CODE",inplace = True)
df.head()
```

```python
print (df.columns)
#Easy copy paste desired columns
```

```python
#Dropping unnecessary columns (for this analysis)
unnecessary = ['MUNICIPALITY', 'CITY_SHIRE','A Crimes against the person 2023',
       'B Property and deception offences 2023', 'C Drug offences 2023',
       'D Public order and security offences 2023',
       'E Justice procedures offences 2023', 'F Other offences 2023']
clean_df = df.drop(columns=unnecessary)
```

```python
#Applying quotient-cut bins to evenly grade our chosen areas of analysis
clean_df["Affordability Livability Grade"] = pd.qcut(df["Median House Price 2022"], q=5, labels = [5,4,3,2,1])
clean_df["Crime/Population Quotient"] = (clean_df["Total Crime 2023"] / clean_df["Estimated_Resident_Population_2021"])
clean_df["Crime Livability Grade"] = pd.qcut(clean_df["Crime/Population Quotient"],q=5,labels=[5,4,3,2,1])
clean_df["School Livability Grade"] = pd.qcut(clean_df["School_Count_2022"],q=3,labels=[1,2,3])

```

```python
#Formatting result for readability
move = clean_df.pop("Crime/Population Quotient")
clean_df.insert(5,"Crime/Population Quotient",move)
clean_df.rename(columns = {"Hospital_Count_2023":"Has Hospital"},inplace = True)
clean_df.rename(columns = {"Police_Count_2023":"Has Police Station"},inplace = True)
clean_df.rename(columns = {"Supermarket_Count_2023":"Has Supermarket"},inplace = True)
clean_df
```

```python
#Calculaing liveability grade across each row
clean_df["Final Liveability Grade"] = clean_df[['Has Hospital',
       'Has Police Station', 'Has Supermarket',
       'Affordability Livability Grade', 'Crime Livability Grade',
       'School Livability Grade']].sum(axis = 1)
```

```python
#Sorting to find best postcodes via this score
sort_df = clean_df.sort_values(by="Final Liveability Grade",ascending = False)
sort_df.head()
```

```python
#Output to csv
sort_df.to_csv('Output/Liveability Analysis Output.csv')
```
