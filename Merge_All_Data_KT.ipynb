# Merge All Data
### Median House Price, Crime, Population, Schools,Hospitals, Police Stations, Supermarkets

```python
# Import the required libraries
import requests
import json
import pandas as pd
import numpy as np
```

```python
#Import house median price Data
#encodings = ['utf-8', 'utf-8-sig', 'latin-1', 'cp1252', 'utf-16']

file_path = "Cleaned_Data/house_by_postcode_2022.csv"
house_df = pd.read_csv(file_path, encoding="utf-8")
# Rename the column
house_df = house_df.rename(columns={"2022": "Median House Price 2022"})

# Set POST_CODE as the index
house_df.set_index('POST_CODE', inplace=True)

# Reformat the "Median House Price 2022" column
house_df["Median House Price 2022"] = house_df["Median House Price 2022"].apply(lambda x: '{:,.0f}'.format(x))

house_df
```

```python
# Set the display format for the specified columns
pd.options.display.float_format = "{:,.0f}".format

# Import crime Data
file_path = "Cleaned_Data/crime_by_postcode_offence_division_2023.csv"
crime_df = pd.read_csv(file_path, encoding="utf-8")

# Rename the columns for 2023
crime_df = crime_df.rename(columns={
    "A Crimes against the person": "A Crimes against the person 2023",
    "B Property and deception offences": "B Property and deception offences 2023",
    "C Drug offences": "C Drug offences 2023",
    "D Public order and security offences": "D Public order and security offences 2023",
    "E Justice procedures offences": "E Justice procedures offences 2023",
    "F Other offences": "F Other offences 2023"
})

# Set POST_CODE as the index
crime_df.set_index('POST_CODE', inplace=True)

crime_df = crime_df.drop(['MUNICIPALITY', 'CITY_SHIRE', 'SUBURB_GROUP','SUBURB_NAMES'], axis=1)
# Display the DataFrame
crime_df
```

```python
#Import population Data
#encodings = ['utf-8', 'utf-8-sig', 'latin-1', 'cp1252', 'utf-16']

file_path = "Cleaned_Data/population_by_postcode_2021.csv"
population_df = pd.read_csv(file_path, encoding="utf-8")
# Rename the column
population_df = population_df.rename(columns={"Estimated_Resident_Population": "Estimated_Resident_Population_2021"})
# Set POST_CODE as the index
population_df.set_index('POST_CODE', inplace=True)

population_df = population_df.drop(['MUNICIPALITY', 'CITY_SHIRE', 'SUBURB_GROUP','SUBURB_NAMES'], axis=1)
population_df
```

```python
#Import school Data
#encodings = ['utf-8', 'utf-8-sig', 'latin-1', 'cp1252', 'utf-16']

file_path = "Cleaned_Data/school_count_postcode_2022.csv"
school_df = pd.read_csv(file_path, encoding="utf-8")

# Set POST_CODE as the index
school_df.set_index('POST_CODE', inplace=True)

# Rename the column
school_df = school_df.rename(columns={"School_Count": "School_Count_2022"})
school_df = school_df.drop(['MUNICIPALITY', 'CITY_SHIRE', 'SUBURB_GROUP','SUBURB_NAMES'], axis=1)
school_df
```

```python
#Import hospital Data
#encodings = ['utf-8', 'utf-8-sig', 'latin-1', 'cp1252', 'utf-16']

file_path = "Cleaned_Data/hospital_count_by_postcode_2023.csv"
hospital_df = pd.read_csv(file_path, encoding="utf-8")
# Set POST_CODE as the index
hospital_df.set_index('POST_CODE', inplace=True)
hospital_df = hospital_df.rename(columns={"Hospital_Count": "Hospital_Count_2023"})
hospital_df = hospital_df.drop(['MUNICIPALITY', 'CITY_SHIRE', 'SUBURB_GROUP'], axis=1)
hospital_df
```

```python
#Import police Data
#encodings = ['utf-8', 'utf-8-sig', 'latin-1', 'cp1252', 'utf-16']

file_path = "Cleaned_Data/police_count_by_postcode_2023.csv"
police_df = pd.read_csv(file_path, encoding="utf-8")
# Set POST_CODE as the index
police_df.set_index('POST_CODE', inplace=True)
police_df = police_df.rename(columns={"Police_Count": "Police_Count_2023"})
police_df = police_df.drop(['MUNICIPALITY', 'CITY_SHIRE', 'SUBURB_GROUP'], axis=1)
police_df
```

```python
#Import supermarket Data
#encodings = ['utf-8', 'utf-8-sig', 'latin-1', 'cp1252', 'utf-16']

file_path = "Cleaned_Data/supermarket_count_by_postcode_2023.csv"
market_df = pd.read_csv(file_path, encoding="utf-8")
# Set POST_CODE as the index
market_df.set_index('POST_CODE', inplace=True)
market_df = market_df.rename(columns={"Supermarket_Count": "Supermarket_Count_2023"})
market_df = market_df.drop(['MUNICIPALITY', 'CITY_SHIRE', 'SUBURB_GROUP'], axis=1)

market_df
```

```python
# Merge the DataFrames
merged_df = house_df.merge(crime_df, on='POST_CODE', how='outer')
merged_df = merged_df.merge(population_df, on='POST_CODE', how='outer')
merged_df = merged_df.merge(school_df, on='POST_CODE', how='outer')
merged_df = merged_df.merge(hospital_df, on='POST_CODE', how='outer')
merged_df = merged_df.merge(police_df, on='POST_CODE', how='outer')
merged_df = merged_df.merge(market_df, on='POST_CODE', how='outer')

# Specify the order of columns
column_order = ['MUNICIPALITY', 'CITY_SHIRE', 'SUBURB_GROUP'] + [col for col in merged_df.columns if col not in ['MUNICIPALITY', 'CITY_SHIRE', 'SUBURB_GROUP']]

# Reorder the columns
merged_df = merged_df[column_order]
merged_df = merged_df.fillna(0)
# Display the merged DataFrame
# Export cleaned population dataset to a CSV file
merged_df.to_csv("Cleaned_Data/merged_all_data.csv", index_label="POST_CODE")
merged_df
```
